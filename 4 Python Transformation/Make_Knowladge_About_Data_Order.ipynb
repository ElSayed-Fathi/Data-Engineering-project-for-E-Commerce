```python
df['order_id'] = df['order_id'].astype("string")
df['user_name'] = df['user_name'].astype("string")
df['order_status'] = df['order_status'].astype("string")
df['order_date'] = df['order_date'].astype("string")
df['order_approved_date'] = df['order_approved_date'].astype("string")
df['pickup_date'] = df['pickup_date'].astype("string")
df['delivered_date'] = df['delivered_date'].astype("string")
df['estimated_time_delivery'] = df['estimated_time_delivery'].astype("string")
```

```python
# Split the 'feedback_answer_date' column into two columns: 'feedback_answer_full_date' and 'feedback_answer_time'
df[['order_date_full', 'order_date_time']] = df['order_date'].str.split(' ', expand=True)
```

```python
df = df.drop('order_date_time', axis=1)
```

```python
df = df.drop('order_date', axis=1)
```

```python
# Split the 'feedback_answer_date' column into two columns: 'feedback_answer_full_date' and 'feedback_answer_time'
df[['order_approved_date_full', 'order_approved_time']] = df['order_approved_date'].str.split(' ', expand=True)
```

```python
df = df.drop('order_approved_time', axis=1)
```

```python
df = df.drop('order_approved_date', axis=1)
```

```python
# Split the 'feedback_answer_date' column into two columns: 'feedback_answer_full_date' and 'feedback_answer_time'
df[['pickup_datefull', 'pickup_date_time']] = df['pickup_date'].str.split(' ', expand=True)
```

```python
df = df.drop('pickup_date_time', axis=1)
```

```python
df = df.drop('pickup_date', axis=1)
```

```python
# Split the 'feedback_answer_date' column into two columns: 'feedback_answer_full_date' and 'feedback_answer_time'
df[['delivered_date_datefull', 'delivered_date_time']] = df['delivered_date'].str.split(' ', expand=True)
```

```python
df = df.drop('delivered_date_time', axis=1)
```

```python
df = df.drop('delivered_date', axis=1)
```

```python
# Split the 'feedback_answer_date' column into two columns: 'feedback_answer_full_date' and 'feedback_answer_time'
df[['estimated_time_delivery_datefull', 'estimated_time_delivery_time']] = df['estimated_time_delivery'].str.split(' ', expand=True)
```

```python
df = df.drop('estimated_time_delivery_time', axis=1)
```

```python
df = df.drop('estimated_time_delivery', axis=1)
```

```python
#Convert the 'feedback_form_sent_date' column from object to date
df['order_date_full'] = pd.to_datetime(df['order_date_full'])
df['order_approved_date_full'] = pd.to_datetime(df['order_approved_date_full'])
df['pickup_datefull'] = pd.to_datetime(df['pickup_datefull'])
df['delivered_date_datefull'] = pd.to_datetime(df['delivered_date_datefull'])
df['estimated_time_delivery_datefull'] = pd.to_datetime(df['estimated_time_delivery_datefull'])
```

```python
# Rename the column
df = df.rename(columns={'order_date_full': 'order_date'})
df = df.rename(columns={'order_approved_date_full': 'order_approved_date'})
df = df.rename(columns={'pickup_datefull': 'pickup_date'})
df = df.rename(columns={'delivered_date_datefull': 'delivered_date'})
df = df.rename(columns={'estimated_time_delivery_datefull': 'estimated_time_delivery'})
```

```python
# Export the DataFrame to a CSV file
df.to_csv('order_dataset_1.csv', index=False)
```
